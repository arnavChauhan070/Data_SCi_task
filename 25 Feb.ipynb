```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("housing_dataset - Sheet1.csv")
df.columns
X = df[['size','bedrooms','bathrooms','age','location']]
Y = df['price']

#splitting
X_train,X_test,Y_train,_Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)
X_train
```

```python
X_test
```

```python
X_train,X_val,Y_train,Y_val = train_test_split(X_train,Y_train,test_size = 0.25 , random_state = 42)

print(f"Trainig Data : {len(X_train)} samples")
print(f"Validation state : {len(X_val)} samples")
print(f"Test Data : {len(X_test)} samples")
```
