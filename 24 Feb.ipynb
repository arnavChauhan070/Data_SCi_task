#  Scenario Question
# A university library tracked the number of books checked out from four sections during the last semester:
# - North Wing: 2,000 books
# - South Wing: 1,500 books
# - East Wing: 1,800 books
# - West Wing: 2,200 books
# The librarian wants to compare which section had the highest activity and which section had the lowest.
# You are asked to:
# - Plot a bar chart showing the number of books checked out by section.
# - Add a title and axis labels to make the chart clear.
# - Interpret the chart:
# - Which section had the highest checkouts?
# - Which section had the lowest?
# - What insights could the librarian use to improve resource allocation?

```python
wings = ["N","S","E","W"]
books = np.array([2,1.5,1.8,2.2]) * 1000
books
plt.bar(wings,books,color = "black")
plt.title("Books read by each seaction")
plt.ylabel("Checkout")
plt.xlabel("Section")
plt.show()
```

```python
highest_checkout = wings[books.argmax()]
highest_checkout
```

```python
# create fig and axis
fig,ax = plt.subplots(figsize = (8,5))

month = ['Jan','Feb','Mar',"April","May","jun"]
sales = [ 42000,55000,49000,67000,71000,83000 ]

ax.plot(month,sales,color = "#00B48B",linewidth = 2.5 , marker = '^' , markersize = 7)

ax.set_title('Monthly Sales 2024 ',fontsize = 14)
ax.set_xlabel('Month')
ax.set_ylabel('Revenue ($)')
ax.grid(axis = 'y',alpha = 0.4)
plt.tight_layout()
plt.show()
```

```python
ad_spend = np.array([2,4,6,8,12,14,16,18]) * 1000
sales = np.array([7,15,21,25,32,36,42,47]) * 1000

plt.scatter(ad_spend,sales,color = 'blue', alpha = 0.7)

z = np.polyfit(ad_spend,sales,1)
p = np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),'r--',label = 'Trend')

plt.xlabel('Ad Spend ($)')
plt.ylabel('Sales ($)')
plt.title('Ad spend Vs sales ')
plt.legend()
plt.show()

```

```python
# how to get prob of tossing coins
import random
head = 0
total = 1000
for i in range(total):
    res = random.randint(0,1)
    if res == 1:
        head+=1
tail = total - head
print("Head",head)
print("Tail",tail)
print("P(Head)",head/total)
print("P(Tail)",tail/total)
```

```python

heights = np.random.normal(165,8,200)
print("Shortest : ",round(heights.min(),1))
print("Tallest : ",round(heights.max(),1))
print("average :",round(heights.mean(),1))

plt.hist(heights,bins = 20,color = "#00BFA6",edgecolor = 'white')
plt.axvline(heights.mean(),color = 'red',linewidth = 2, label = "Average")

```

```python
#binomial distribution
from scipy.stats import binom
n = 10 
p = 0.25

p3 = binom.pmf(3,n,p)
print("P(exactly 3 ) = ",round(p3,3))

p5 = 1 - binom.pmf(4,n,p)
print("P(exactly 5 ) = ",round(p5,3))

```
