Which department has max emp

```python
max_val = 0
max_dept = ""
l = [emp_data,sales_data]
for idx in l:
    for dept in idx["Department"]:
        if len(dept) > max_val:
            max_val = len(dept)
            max_dept = dept
    print("Max Department",max_dept)
```
