```python

import pandas as pd
import numpy as np

customers = pd.read_csv("Customers.csv")
sales = pd.read_csv("Sales.csv")   
support = pd.read_csv("Support.csv")

print("\nCustomers Info:")
print(customers.info())
print("\nSales Info:")
print(sales.info())
print("\nSupport Info:")
print(support.info())

customers["Age"].fillna(customers["Age"].median(), inplace=True)

customers.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)
sales.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)
support.rename(columns={"CustomerID": "Cust_ID"}, inplace=True)

sales["DiscountedPrice"] = sales["Price"] * 0.9

# Compute revenue per order
sales["Revenue"] = sales["Quantity"] * sales["Price"]

april_orders = sales[sales["OrderDate"].str.startswith("2023-04")]
print("\nApril Orders:\n", april_orders)

# Slice first 5 rows of sales
print("\nFirst 5 Sales Records:\n", sales.head())

north_customers = customers[customers["Region"] == "North"]

high_value_orders = sales[sales["Revenue"] > 50000]

print("\nNorth Customers:\n", north_customers)
print("\nHigh Value Orders:\n", high_value_orders)


sorted_customers = customers.sort_values(by="SignupDate")

sorted_sales = sales.sort_values(by="Revenue", ascending=False)

print("\nSorted Customers:\n", sorted_customers)
print("\nSorted Sales:\n", sorted_sales)


avg_revenue_product = sales.groupby("Product")["Revenue"].mean()

avg_resolution_issue = support.groupby("IssueType")["ResolutionTime"].mean()

print("\nAverage Revenue per Product:\n", avg_revenue_product)
print("\nAverage Resolution Time per Issue Type:\n", avg_resolution_issue)


merged_data = sales.merge(customers, on="Cust_ID").merge(support, on="Cust_ID")

print("\nMerged Dataset:\n", merged_data.head())


clv = sales.groupby("Cust_ID")["Revenue"].sum().reset_index()
clv.rename(columns={"Revenue": "CustomerLifetimeValue"}, inplace=True)

customers = customers.merge(clv, on="Cust_ID", how="left")

print("\nCustomers with CLV:\n", customers)

merged_data.to_csv("Cleaned_Data.csv", index=False)
print("\nCleaned dataset exported to Cleaned_Data.csv")

```

```python
import pandas as pd
import numpy as np

# Load data
customers_df = pd.read_csv("Customers1 - Sheet1.csv")
sales_df = pd.read_csv("Sales1 - Sheet1.csv")
support_df = pd.read_csv("support1 - Sheet1.csv")

# Basic inspection
print(customers_df.columns)
print(sales_df.columns)
print(support_df.columns)

# Discount and revenue calculation
sales_df["Discounted_Price"] = sales_df["Price"] * 0.90
sales_df["Revenue"] = sales_df["Quantity"] * sales_df["Price"]

# Convert dates
sales_df["OrderDate"] = pd.to_datetime(sales_df["OrderDate"])
customers_df["SignupDate"] = pd.to_datetime(customers_df["SignupDate"])

# Orders in January 2025
jan_2025_orders = sales_df[
    sales_df["OrderDate"].dt.to_period("M") == "2025-01"
]

# First 10 rows
first_10_sales = sales_df.head(10)

# Filtering
north_customers = customers_df[customers_df["Region"] == "North"]
high_revenue_orders = sales_df[sales_df["Revenue"] > 10000]

# Sorting
customers_sorted = customers_df.sort_values("SignupDate")
sales_sorted = sales_df.sort_values("Revenue", ascending=False)

# Grouping
sales_with_region = pd.merge(
    sales_df,
    customers_df[["CustomerID", "Region"]],
    on="CustomerID",
    how="left"
)

avg_revenue_by_region = (
    sales_with_region.groupby("Region")["Revenue"].mean().reset_index()
)

avg_resolution_by_issue = (
    support_df.groupby("IssueType")["ResolutionTime"].mean().reset_index()
)

# Handle missing age
customers_df["Age"] = customers_df["Age"].fillna(customers_df["Age"].median())

# Merge datasets
merged_df = pd.merge(customers_df, sales_df, on="CustomerID", how="left")
full_df = pd.merge(merged_df, support_df, on="CustomerID", how="left")

# CLV
clv = sales_df.groupby("CustomerID")["Revenue"].sum().reset_index()
clv.rename(columns={"Revenue": "CLV"}, inplace=True)

full_df = pd.merge(full_df, clv, on="CustomerID", how="left")

# Average resolution per customer
avg_res_customer = (
    support_df.groupby("CustomerID")["ResolutionTime"].mean().reset_index()
)
avg_res_customer.rename(
    columns={"ResolutionTime": "AvgResolutionTime"},
    inplace=True
)

full_df = pd.merge(full_df, avg_res_customer, on="CustomerID", how="left")

# Export
full_df.to_csv("Cleaned_Data.csv", index=False)
```
